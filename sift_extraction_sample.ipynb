```python
def show_lab(li,index):
    lab_list=[]
    for file in li:
        mat_file=io.loadmat(os.path.join(os.getcwd(),'ClothingAttributeDataset/labels/',file))
        lab_list.append(mat_file['GT'][index])
    return lab_list
```

```python
#show_lab 함수를 통해 각 사진의 26개의 label을 볼 수 있음
show_lab(lab_list,0)
```
